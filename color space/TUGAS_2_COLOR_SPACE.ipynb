```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
from matplotlib import pyplot as plt 
```

Membuat variabel Gambar

```python
spiral = cv2.imread("/content/drive/MyDrive/Colab Notebooks/sample_images/rgb_image.jpg")
```

Menampilkan Gambar

```python
cv2_imshow(spiral)
```

Dimensi Gambar

```python
print(spiral.shape)
```

Konversi Warna

```python
spiral_abu = cv2.cvtColor(spiral, cv2.COLOR_BGR2GRAY)
spiral_rgb = cv2.cvtColor(spiral, cv2.COLOR_BGR2RGB)
spiral_hsv = cv2.cvtColor(spiral, cv2.COLOR_BGR2HSV)
```

```python
cv2_imshow(spiral_abu)
print(spiral_abu.shape)
cv2_imshow(spiral_rgb)
print(spiral_rgb.shape)
cv2_imshow(spiral_hsv)
print(spiral_hsv.shape)
```

Memisahkan channel warna

```python
b,g,r = cv2.split(spiral)

cv2_imshow(b)
print(b.shape)

cv2_imshow(g)
print(g.shape)

cv2_imshow(r)
print(r.shape)
```

```python
gabung_bgr = cv2.merge([b,g,r])

gabung_rgb = cv2.merge([r,g,b])

cv2_imshow(gabung_bgr)
cv2_imshow(gabung_rgb)
```
